```python
import datapane as dp
import pandas as pd
import plotly.express as px
import random
```

```python
def generate_doors(rng, num_doors):
    doors = [0 for _ in range(num_doors)]
    doors[0] = 1
    rng.shuffle(doors)
    return doors
```

```python
def choose_door(rng, num_doors):
    return rng.choice(range(num_doors))
```

```python
def choose_unopened_door(rng, doors, choice):
    if doors[choice] == 1:
        idx = list(range(len(doors)))
        idx.remove(choice)
        return rng.choice(idx)
    return doors.index(1)
```

```python
def run_sim(rng, n, num_doors):
    results_df = pd.DataFrame({"prize": [], "choice": [], "unopened": []}, dtype=int)
    for _ in range(n):
        doors = generate_doors(rng, num_doors)
        choice = choose_door(rng, num_doors)
        unopened = choose_unopened_door(rng, doors, choice)
        results_df = results_df.append({
            "prize": doors.index(1),
            "choice": choice, 
            "unopened": unopened
        }, ignore_index=True)
    results_df["chose_prize"] = results_df["prize"] == results_df["choice"]
    return results_df
```

```python
n = 1000
rng = random.Random(42)
results = run_sim(rng, n, 3)
results
```

```python
fig = px.bar(
    results.groupby("chose_prize").count().reset_index(), 
    x="chose_prize", 
    y="prize",
    labels = {
        "chose_prize": "Whether the chosen door had the prize",
        "prize": "Count",
    },
    title="Monty Hall Problem Simulation with 3 Doors",
)
fig.show()
```

```python
dp.Report(dp.Plot(fig)).upload(name="Monty Hall Problem Simulation - 3 Doors")
```
